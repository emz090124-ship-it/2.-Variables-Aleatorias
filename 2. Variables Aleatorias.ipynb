### Introducción a las Variables Aleatorias

Las variables aleatorias son funciones que asignan un valor numérico a cada resultado de un experimento aleatorio. Este concepto permite representar fenómenos inciertos de forma cuantitativa, facilitando su análisis mediante herramientas matemáticas y estadísticas.

Existen dos tipos principales:

- **Variable aleatoria discreta**: toma valores aislados, generalmente enteros.  
  Ejemplo: el número de caras al lanzar tres monedas.

- **Variable aleatoria continua**: toma valores dentro de un intervalo o conjunto no numerable.  
  Ejemplo: el tiempo exacto que tarda un cliente en ser atendido.

El estudio de variables aleatorias incluye conceptos como:

- Función de probabilidad o densidad
- Función de distribución acumulada
- Esperanza matemática (valor esperado)
- Varianza y desviación estándar

Este marco es fundamental para modelar fenómenos en economía, física, ingeniería y ciencias sociales.

Las variables aleatorias discretas toman valores numéricos aislados, generalmente enteros, y se utilizan para modelar fenómenos donde los resultados posibles son contables. En esta sección se presentan varias **familias paramétricas de distribuciones discretas univariadas**, incluyendo sus funciones de probabilidad, medias, varianzas y ejemplos de experimentos aleatorios que pueden modelarse con ellas.

Cada familia está definida por uno o más **parámetros**, que determinan la forma y comportamiento de la distribución. No existe una notación uniforme para los parámetros: se utilizan tanto letras griegas como latinas.

#### Distribución Uniforme Discreta

La **distribución uniforme discreta** es aquella en la que todos los valores posibles tienen la misma probabilidad. Se define sobre un conjunto finito de enteros consecutivos.

**Definición:**  
Sea $x \in \{1, 2, \dots, N\} $, donde $N$ es un entero positivo. La función de masa de probabilidad se define como:


$$
  f(x) = \frac{1}{N}, \quad \text{para } x = 1, 2, \dots, N
$$

y  


$
f(x) = 0 \quad \text{en cualquier otro caso}.
$



Esta distribución es útil para modelar situaciones donde cada resultado tiene la misma probabilidad de ocurrir, como seleccionar un número al azar entre 1 y $N$.

> Ejemplo:  
> Al lanzar un dado justo de seis caras, la variable aleatoria que representa el número obtenido sigue una distribución uniforme discreta con $(N = 6)$.

---


### Función de Masa y Propiedades

La distribución uniforme discreta asigna la misma probabilidad a cada uno de los (N) valores posibles. Es uno de los modelos más simples y útiles para representar incertidumbre cuando no hay razón para favorecer un resultado sobre otro.

#### Función de Masa de Probabilidad

Sea $X$ una variable aleatoria con distribución uniforme discreta sobre el conjunto $\{1, 2, \dots, N\}$. Su función de masa de probabilidad (fmp) está dada por:

$$
  f(x) = \mathbb{P}(X = x) = 
  \begin{cases}
    \frac{1}{N}, & \text{si } x \in \{1, 2, \dots, N\} \\
    0, & \text{en otro caso}
  \end{cases}
$$

#### Propiedades

- **Espacio muestral**: $\{1, 2, \dots, N\}$
- **Media (esperanza matemática)**:
  
$$
  \mathbb{E}[X] = \frac{N + 1}{2}
$$

- **Varianza**:
$$
\text{Var}(X) = \frac{N^2 - 1}{12}
$$



> Fuente: Mood, A. M., Graybill, F. A., & Boes, D. C. (2007). *Introducción a la teoría de la probabilidad*. Sección 2.1, Distribución uniforme discreta. Facultad de Ciencias, UNAM.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson, nbinom, hypergeom
plt.style.use('seaborn-v0_8')

N = 6
x = np.arange(1, N+1)
p = np.ones(N) / N

plt.bar(x, p, color='skyblue')
plt.title("Distribución Uniforme Discreta U{1,...,6}")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

print(f"Esperanza = {(N+1)/2}")
print(f"Varianza = {(N**2 - 1)/12}")
```

## Variable aleatoria Bernoulli con parámetro $p\in (0,1)$.

El modelo probabilístico Bernoulli se aplica a un experimento cuyo espacio muestral está constituido sólo por dos resultados posibles, éxito y fracaso:

Se considerará una v.a $X$ sobre el espacio muestral $$\Omega=\{\text{éxito, fracaso}\}$$
de tal forma que
* $X(\{\text{éxito}\}) = 1$
* $X(\{\text{fracaso}\}) = 0$

Las probabilidades asociadas a este modelo son:

* $\mathbb{P}(\{\text{éxito}\}) = p$
* $\mathbb{P}(\{\text{fracaso}\}) = 1-p$

donde $0<p<1$.


Entonces la función de densidad de masa de probabilidades de $X$ está dada por:

$$f_{X}(k)=\begin{cases}
p &\quad \text{si } k=1 \\
1-p &\quad \text{si } k=0
\end{cases}$$

**Notación:** $X\sim \text{Ber}(p)$.

Esta variable aleatoria es muy importante. En la practica es usada para modelar situaciones en las cuales hay dos posibles resultados como:
* el estado de un teléfono en un momento dado: libre u ocupado.
* una persona pueda estar enferma o sana de una determinada enfermedad.
* la preferencia de una persona, la cual puede estar a favor o en contra de un candidato determinado.

Mediante la combinación de variables aleatorias Bernoulli es posible construir otras variables aleatorias.

### Esperanza y Varianza

Si $X\sim \text{Ber}(p)$, entonces su esperanza es:
\begin{align*}
\mathbb{E}[X] = \sum_{k=0}^{1}kf_{X}(k) = 1\cdot p + 0\cdot (1-p) = p.
\end{align*}

Y su varianza se calcula usando,
$$\mathbb{E}[X^{2}] = \sum_{k=0}^{1}k^{2}f_{X}(k) = 1^{2}\cdot p + 0^{2}\cdot (1-p) = p.$$
Por lo tanto,
$$\text{Var}(X)=\mathbb{E}[X^{2}]-(\mathbb{E}[X])^{2} = p-p^{2} = p(1-p)$$

```python
p = 0.4
x = [0, 1]
pmf = bernoulli.pmf(x, p)

plt.bar(x, pmf, color='salmon')
plt.title("Distribución Bernoulli (p=0.4)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

print(f"Esperanza = {p}")
print(f"Varianza = {p*(1-p)}")

```

## Variable aleatoria Binomial con parámetros $n$ y $p\in (0,1)$.

Decimos que una v.a $X$ tiene distribución binomial con párametros $n$ y $p$, si su función de densidad de masa de probabilidades está dada por:
$$\begin{equation*}
	f_{X}(k)=\begin{cases}
	\binom{n}{k}p^{k}(1-p)^{n-k} &\quad \text{si } k=0,1,\dots,n \\
	\qquad 0 &\quad \text{en otro caso}  
	\end{cases}
\end{equation*}$$
donde $n$ es un entero positivo y $0< p< 1$.

La esperanza de $X$ es:
\begin{align*}
\mathbb{E}[X]&= 
&=np
\end{align*}

La varianza de $X$ es:
$$\text{Var}(X)=np(1-p).$$


**Ejemplo** Una moneda no justa se tira 6 veces, donde la probabilidad de obtener sol es de $0.3$.

Sea $X$ el número de veces que cae sol

entonces $X$ tiene una distribución binomial con parámetros $n=6$ y $p=0.3$. Por lo que
$$\begin{align*}
\mathbb{P}(X=2) &= \binom{6}{2}(0.3)^{2}(0.7)^{4}=0.3241 \\
\mathbb{P}(X=3) 
			&= 0.5093 + 0.0595 + 0.0102 = 0.579
\end{align*}$$

```python
n, p = 10, 0.3
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='lightgreen')
plt.title("Distribución Binomial (n=10, p=0.3)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

print(f"Esperanza = {n*p}")
print(f"Varianza = {n*p*(1-p)}")
```

## Variable aleatoria Binomial

La distribución binomial modela el **número de éxitos en n ensayos independientes**, cada uno con probabilidad de éxito \( p \).  
Es una extensión del modelo Bernoulli.

### Función de Masa
$$
f(x) = \binom{n}{x} p^x (1-p)^{n-x}
$$

**Esperanza:** $$\mathbb{E}[X] = np $$  
**Varianza:** $$ Var(X) = np(1-p) $$


```python
p = 0.4
x = [0, 1]
pmf = bernoulli.pmf(x, p)

plt.bar(x, pmf, color='salmon')
plt.title("Distribución Bernoulli (p=0.4)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

print(f"Esperanza = {p}")
print(f"Varianza = {p*(1-p)}")
```

## Variable aleatoria Geométrica con parámetro $p\in (0,1)$.

Esta variable aleatoria cuenta el número de fracasos antes del primer éxito en ensayos bernoulli independientes con parámetro $0<p<1$, y su función de masa de probabilidades está dada por:
$$f_{X}(k)=\begin{cases}
	p(1-p)^{k-1} &\quad \text{si } k=1,2,\dots \\
	\qquad 0 &\quad \text{en otro caso}  
\end{cases}$$

La esperanza de $X$ es:
\begin{align*}
\mathbb{E}[X]&=
&=\frac{1}{p}
\end{align*}

**Varianza:** $$Var(X) = \frac{1-p}{p^2} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom


# El parametro de la distribución geométrica 
p = 0.1

# Valores posibles (desde 1 hasta 20 ensayos)
x = np.arange(1, 21)

pmf = geom.pmf(x, p)

plt.figure(figsize=(12,8))

plt.stem(x, pmf, basefmt=" ") # Grafica de barras verticales
plt.title('Distribución Geométrica (p = {:.2f})'.format(p))
plt.xlabel('Número de ensayos hasta el primer éxito')
plt.ylabel('Probabilidad')
plt.show()
```

## Variable aleatoria Poisson

Es una distribución de probabilidad discreta que  sirve para calcular la probabilidad de que ocurra un determinado número de eventos raros durante un intervalo dado (puede ser de tiempo, longitud, área, etc.

Esta variable aleatoria toma los valores sobre el conjunto $\{0,1,2,\dots\}$ y tiene un parámetro $\lambda>0$, el cual representa el número de veces que se espera que ocurra un evento durante un intervalo dado.

La función de masa de probabilidades de una variable aleatoria Poisson, $X$, está dada por:
$$f_{X}(k)=\begin{cases}
	e^{-\lambda}\frac{\lambda^{k}}{k!} &\quad \text{si } k=0,1,2,\dots \\
	0 &\quad \text{en otro caso}  
\end{cases}$$

**Esperanza:**  $$\mathbb{E}[X] = \lambda $$ 
**Varianza:** $$ Var(X) = \lambda $$

```python
lmbda = 3
x = np.arange(0, 10)
pmf = poisson.pmf(x, lmbda)

plt.bar(x, pmf, color='plum')
plt.title("Distribución Poisson (λ=3)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

print(f"Esperanza = {lmbda}")
print(f"Varianza = {lmbda}")
```

## Variable aleatoria Binomial Negativa con parámetros $r\geq 1$ y $p\in (0,1)$.

Supongamos que se realizan ensayos independientes, cada uno con probabilidad $0<p<1$ de ser un éxito, hasta obtener un total de $r$ éxitos acumulados. Sea $X$ el número de ensayos que se requieren, entonces su función de masa de probabilidades está dada por:
$$f_{X}(k)=\begin{cases}
	\binom{k-1}{r-1}p^{r}(1-p)^{k-r} &\quad \text{si } k=r,r+1,\dots \\
	\qquad 0 &\quad \text{en otro caso}  
\end{cases}$$



Notación. $X\sim \text{BN}(r,p)$.

Se tiene que
$$\mathbb{E}[X]=\frac{r}{p}$$
y
$$\text{Var}[X]=\frac{r(1-p)}{p^2}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sps
import random 

# Parámetros de la distribución binomial negativa 
r = 50 # Número de éxitos 
p = 0.2 # Probabilidad de éxito 
s = 100000 # Número de muestras

random.seed(3)

# Generamos números aleatorios con distribución binomial negativa
nbn_numeros = sps.nbinom.rvs(r, p, size=s)

# Generamos el histograma
plt.figure(figsize=(10, 5))
plt.hist(
    nbn_numeros,
    bins=30,
    density=True,
    color='g',
    edgecolor='black'
)
plt.title('Distribución Binomial Negativa')
plt.xlabel('Número de fracasos antes del r-ésimo éxito')
plt.ylabel('Probabilidad')
plt.show()
```

## Variable aleatoria Hipergeométrica con parámetros $n,N,m$.

Supongamos que se elige, sin reemplazo, una muestra de tamaño $n$ de una urna que contiene $N$ bolas, de las cuales $m$ son rojas y $N-m$ son verdes. Sea $X$ el número de ebolas rojas seleccionadas, entonces su función de masa de probabilidades está dada por:
$$f_{X}(i)=
	\frac{\binom{m}{i}\binom{N-m}{n-i}}{\binom{N}{n}} \quad \text{si } i=0,1,\dots, n  
$$



Notación. $X\sim \text{Hiper}(n,N,m)$.

Se tiene que
$$\mathbb{E}[X]=\frac{nm}{N}$$
y
$$\text{Var}[X]=\frac{nm}{N}\left[\frac{(n-1)(m-1)}{N-1}+1-\frac{nm}{N} \right].$$

```python
N, K, n = 50, 10, 5
x = np.arange(0, n+1)
pmf = hypergeom.pmf(x, N, K, n)

plt.bar(x, pmf, color='mediumpurple')
plt.title("Distribución Hipergeométrica (N=50, K=10, n=5)")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()

esperanza = n * (K/N)
varianza = n * (K/N) * ((N-K)/N) * ((N-n)/(N-1))
print(f"Esperanza = {esperanza:.2f}")
print(f"Varianza = {varianza:.2f}")
```

  <h1>Variables Aleatorias Continuas</h1>
</div>

**Definición.** Una variable aleatoria absolutamente continua es aquella para la cual existe una función $f_X$ no negativa, llamada función de densidad, tal que:
$$
F_X(x) = \mathbb{P}(X\leq x) = \int_{-\infty}^{x} f_X(y) dy, \quad x\in \mathbb{R}
$$


---

## Propiedades de las Variables Aleatorias Continuas

Una **variable aleatoria continua** puede tomar cualquier valor dentro de un intervalo real. Sus propiedades principales son:

1. **Función de densidad de probabilidad (f.d.p.)**  
   $$f_X(x) \ge 0$$  $$para \ todo$$   $$x , y$$  
   
   $$\int_{-\infty}^{\infty} f_X(x)\,dx = 1
   $$

2. **Probabilidad en un intervalo:**  
   $$
   P(a \le X \le b) = \int_a^b f_X(x)\,dx
   $$

3. **Valor esperado:**  
   $$
   \mathbb{E}[X] = \int_{-\infty}^{\infty} x f_X(x)\,dx
   $$

4. **Varianza:**  
   $$
   \text{Var}(X) = \int_{-\infty}^{\infty} (x - \mathbb{E}[X])^2 f_X(x)\,dx
   $$

## Variable aleatoria uniforme sobre el intervalo $(a,b)$.

Una variable aleatoria $X$ se dice que tiene distribución uniforme continua  en el intervalo $(a,b)$ con $a,b\in\mathbb{R}$, si su función de densidad esta dada por:
$$f_{X}(x)=\begin{cases}
	\frac{1}{b-a} &\quad \text{si } a< x < b\\
	\qquad 0 &\quad \text{en otro caso}  
\end{cases}$$

La función de distribución de $X\sim Unif((a,b))$ esta dada por:
$$F_{X}(x)=\begin{cases}
	 0 &\quad \text{si } x\le a \\
	\frac{x-a}{b-a} &\quad \text{si } a<x<b\\
	 1 &\quad \text{si } x\ge b  
\end{cases}$$

La esperanza de una variable aleatoria uniforme es la siguiente,
$$\begin{align*}
\mathbb{E}[X] &=  \frac{a+b}{2}.
\end{align*}$$

La varianza es la siguiente

$$Var(X) = \frac{(b-a)^{2}}{12}.$$

```python
from scipy.stats import uniform

a, b = 2, 6
x = np.linspace(0, 8, 200)
pdf = uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, pdf, color='skyblue', lw=2)
plt.fill_between(x, pdf, color='skyblue', alpha=0.3)
plt.title("Distribución Uniforme Continua U(2, 6)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

print(f"Esperanza = {(a+b)/2}")
print(f"Varianza = {((b-a)**2)/12}")
```

# Variable aleatoria normal con parámetros media $\mu$ y varianza $\sigma^{2}$.

La distribución normal es una de las más importantes y de mayor uso tanto en la teoría de la probabilidad, como en la teoría estadística.

También llamada distribución gaussiana, en honor a Gauss, a quien se considera el padre de ésta distribución.

La importancia de la distribución normal, radica en el famoso Teorema central del límite. Fue descubierta por De Moivre en 1733 como un límite de la distribución binomial.

Modela fenómenos naturales donde los valores tienden a concentrarse alrededor de la media.

Se dice que la variable aleatoria $X$ tiene distribución normal de parámetros $\mu$ y $\sigma^{2}$, donde $\mu,\sigma\in\mathbb{R}$ y $\sigma>0$, si su función de densidad está dada por:
$$f_{X}(x)=\begin{cases}
	\frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}} &\quad \text{si }  x \in\mathbb{R} \\
	\qquad 0 &\quad \text{e.o.c}  
\end{cases}$$


**Esperanza:** $$\mathbb{E}[X] = \mu$$  
**Varianza:** $$Var(X) = \sigma^2$$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  

# Parámetros de la normal estándar
mu, sigma = 0, 1
x = np.linspace(-4, 4, 400)
pdf = norm.pdf(x, mu, sigma)

# Visualización
plt.plot(x, pdf, color='salmon', lw=2)
plt.fill_between(x, pdf, color='salmon', alpha=0.3)
plt.title("Distribución Normal N(0,1)", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()

# Propiedades
print(f"Esperanza = {mu}")
print(f"Varianza = {sigma**2}")
```

## Variable aleatoria exponencial con parámetro $\lambda >0$.

La distribución exponencial es una de las distribuciones continuas más utilizadas. A menudo se utiliza para modelar el tiempo transcurrido entre eventos.  

Algunos ejemplos en los que podría utilizarse la distribución exponencial son:
* El tiempo transcurrido en un call center hasta recibir la primer llamada del día.
* El  tiempo entre terremotos de una determinada magnitud.
* Supongamos una máquina que produce hilo de alambre, la cantidad de metros de alambre hasta encontrar una falla en el alambre se podría modelar como una exponencial.

Se dice que la variable aleatoria $X$ tiene distribución exponecial de parámetro $\lambda>0$, si su función de densidad está dada por:
$$f_{X}(x)=\begin{cases}
\lambda e^{-\lambda x} & x>0\\
0 & \text{en otro caso}
\end{cases}$$

La esperanza y varianza de de una distribución exponencial de la forma $$f_{X}(x)=\begin{cases}
\frac{1}{\lambda} e^{-\frac{x}{\lambda}} & x>0\\
0 & \text{en otro caso}
\end{cases}$$
$$\mathbb{E}[X] = \lambda$$
y
\begin{align*}
\text{Var}(X) &=  \lambda^{2}
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon  

# Parámetro de la distribución exponencial
lmbda = 0.5
x = np.linspace(0, 10, 400)

# Función de densidad y distribución acumulada
pdf = expon.pdf(x, scale=1/lmbda)
cdf = expon.cdf(x, scale=1/lmbda)

# Visualización
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Densidad
axs[0].plot(x, pdf, color='orange', lw=2)
axs[0].fill_between(x, pdf, color='orange', alpha=0.3)
axs[0].set_title("Función de Densidad Exponencial (λ = 0.5)", fontsize=13)
axs[0].set_xlabel("x")
axs[0].set_ylabel("f(x)")
axs[0].grid(True, linestyle=':', alpha=0.5)

# Distribución acumulada
axs[1].plot(x, cdf, color='teal', lw=2)
axs[1].set_title("Función de Distribución Acumulada (FDA)", fontsize=13)
axs[1].set_xlabel("x")
axs[1].set_ylabel("F(x)")
axs[1].grid(True, linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()

# Propiedades teóricas
print(f"Esperanza = {1/lmbda}")
print(f"Varianza = {1/(lmbda**2)}")

```

## Variable aleatoria Gamma con parámetros $\alpha$ y $\lambda$.

 La distribución gamma se obtiene al considerar el tiempo que transcurre entre cierto número de ocurrencias de eventos que ocurren aleatoriamente en el tiempo

La función gamma $\Gamma:(0,\infty)\to \mathbb{R}$ está definida como
	$$\Gamma(\alpha)=\int_{0}^{\infty}t^{\alpha-1}e^{-t}dt.$$



*Propiedades de la función gamma*.
- $\Gamma(\alpha)<\infty$ para cualquier $\alpha>0$.
- $\Gamma(\alpha+1)=\alpha\Gamma(\alpha)$.
- Si $n\geq 1$ $\Gamma(n)=(n-1)!$
- $\Gamma\left(\frac{1}{2} \right)=\sqrt{\pi}$.

## Esperanza y Varianza 

La esperanza, $\text{E}[X]$ para una variable aleatoria $X\sim Γ(\alpha, \lambda)$ es:

$$
\mathbb{E}(X) = \frac{\alpha}{\lambda}
$$
y su varianza es:
$$
\text{Var}(X) = \frac{\alpha}{\lambda^2}
$$



La distribución gamma es esencial en varios campos por su capacidad para modelar tiempos de espera y eventos con tasas constantes. Sus aplicaciones incluyen:

  * $\textbf{Teoría de colas y procesos estocásticos:}$ Modela el tiempo de espera hasta el k-ésimo evento.
  * $\textbf{Confiabilidad y análisis de supervivencia:}$ Utilizada para tiempos de fallo y eventos críticos en medicina.
  * $\textbf{Hidrología y meteorología:}$ Aplica en la modelación de precipitaciones acumuladas y tamaños de gotas.
  * $\textbf{Procesamiento de imágenes y señales:}$ En el ajuste de modelos a datos de intensidades.
  * $\textbf{Finanzas:}$ Para rendimientos de activos que no siguen distribuciones normales.
  * $\textbf{Biología y ecología:}$ Para tasas de crecimiento y tiempos entre eventos biológicos.
  * $\textbf{Física:}$ Describe tiempos de decaimiento y distribuciones de energía.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma  # ← Importación necesaria

# Parámetros de la distribución Gamma
alpha, beta = 2, 2
x = np.linspace(0, 20, 400)
pdf = gamma.pdf(x, a=alpha, scale=beta)

# Visualización
plt.plot(x, pdf, color='purple', lw=2)
plt.fill_between(x, pdf, color='purple', alpha=0.3)
plt.title("Distribución Gamma (α = 2, β = 2)", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()

# Propiedades teóricas
print(f"Esperanza = {alpha * beta}")
print(f"Varianza = {alpha * (beta ** 2)}")
```

# Simulación de Variables Aleatorias y la Transformada Inversa

En estadística y probabilidad, la **simulación de variables aleatorias** permite generar muestras de una distribución específica usando métodos computacionales. Una de las técnicas más comunes es la **transformada inversa**, que se basa en la siguiente idea:

> Si $ U \sim \text{Uniforme}(0,1) $ y $F_X$ es la función de distribución acumulada (FDA) de una variable aleatoria $X$, entonces
> 
> 
> $$X = F_X^{-1}(U)$$
> 
> 
> tiene la distribución deseada $X$.

---

## Paso a paso del método de la transformada inversa

1. Obtener la **función de distribución acumulada** $F_X(x)$ de la variable aleatoria deseada.
2. Generar números aleatorios $U_1, U_2, \dots, U_n \sim \text{Uniforme}(0,1)$.
3. Aplicar la **función inversa** $F_X^{-1}(U_i)$ para obtener valores simulados de $X$.

Este método funciona para distribuciones continuas donde se puede calcular $F_X^{-1}$ de manera analítica.

---

## Ejemplo: Distribución Exponencial

La distribución **exponencial** con parámetro $\lambda > 0$ tiene la función de distribución acumulada:

$$
F_X(x) = 1 - e^{-\lambda x}, \quad x \ge 0
$$

Su inversa es:

$$
F_X^{-1}(u) = -\frac{1}{\lambda} \ln(1-u)
$$

Generando números uniformes $U \sim \text{Uniforme}(0,1)$ y aplicando la transformada inversa, obtenemos valores simulados de la variable aleatoria exponencial.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial
lam = 0.5
n = 1000  # número de simulaciones

# Paso 1: generar números uniformes
u = np.random.uniform(0, 1, n)

# Paso 2: aplicar la transformada inversa
x = -np.log(1 - u) / lam

# Visualización: histograma de los valores simulados
plt.hist(x, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title("Simulación de variable aleatoria Exponencial (λ=0.5)")
plt.xlabel("x")
plt.ylabel("Frecuencia relativa")
plt.show()
```

# Funciones Generadoras

En probabilidad, las **funciones generadoras** permiten resumir la información de una variable aleatoria y obtener sus momentos (media, varianza, etc.) de manera sistemática. Las más comunes son:

---

## 1. Función Generadora de Momentos (FGM)

La **función generadora de momentos** (MGF, por sus siglas en inglés) de una variable aleatoria $X$ se define como:

$$
M_X(t) = \mathbb{E}[e^{tX}] = \int_{-\infty}^{\infty} e^{tx} f_X(x) dx \quad \text{(para continua)}
$$

$$
M_X(t) = \sum_{x} e^{tx} P(X=x) \quad \text{(para discreta)}
$$

**Propiedades importantes:**
1. $M_X(0) = 1$
2. Si $M_X(t)$ existe en un entorno de $t=0$, los momentos se obtienen derivando:
   $$
   \mathbb{E}[X^n] = M_X^{(n)}(0)
   $$
   donde $M_X^{(n)}(0)$ es la n-ésima derivada de $M_X(t)$ evaluada en $t=0$.
3. Si dos variables tienen la misma FGM en un entorno de $t=0$, tienen la misma distribución.

**Ejemplo de momentos:**
- Media: $\mathbb{E}[X] = M_X'(0)$
- Varianza: $\text{Var}(X) = M_X''(0) - (M_X'(0))^2$

---

## 2. Función Característica (FC)

La **función característica** de $X$ se define como:

$$
\phi_X(t) = \mathbb{E}[e^{itX}] \quad , \quad i = \sqrt{-1}
$$

**Propiedades importantes:**
1. $\phi_X(0) = 1$
2. Siempre existe para cualquier variable aleatoria.
3. Los momentos se pueden obtener de manera similar:
   $$
   \mathbb{E}[X^n] = \frac{1}{i^n} \phi_X^{(n)}(0)
   $$
4. La FC determina de manera única la distribución de $X$.

**Relación con MGF:**  
Si $M_X(t)$ existe para $t$ real, entonces
$$
\phi_X(t) = M_X(it)
$$

---

## Ejemplo: Variable aleatoria Poisson $X \sim \text{Poisson}(\lambda)$

**FGM:**  
$$
M_X(t) = \exp(\lambda(e^t - 1))
$$

- Media: $M_X'(0) = \lambda$
- Varianza: $M_X''(0) - (M_X'(0))^2 = \lambda$

**FC:**  
$$
\phi_X(t) = \exp(\lambda(e^{it} - 1))
$$

```python
import numpy as np
from scipy.stats import poisson

# Parámetro Poisson
lam = 3
n = 10000

# Simulación de variable aleatoria Poisson
X = poisson.rvs(mu=lam, size=n)

# Estimación de momentos a partir de la simulación
media_est = np.mean(X)
varianza_est = np.var(X)

print(f"Media estimada: {media_est:.2f}")
print(f"Varianza estimada: {varianza_est:.2f}")

# Función generadora de momentos teórica para Poisson
def mgf_poisson(t, lam):
    return np.exp(lam * (np.exp(t) - 1))

# Ejemplo: MGF y derivadas (aproximación numérica)
t = 0.0
mgf_val = mgf_poisson(t, lam)
print(f"MGF en t=0: {mgf_val:.2f}")
```

# Ley de los Grandes Números (LGN)

La **Ley de los Grandes Números** es un resultado fundamental en probabilidad que describe el comportamiento promedio de una secuencia de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) a medida que aumenta el número de observaciones.

---

## Definición

Sea $X_1, X_2, \dots, X_n$ una secuencia de variables aleatorias i.i.d. con **media finita** $\mu = \mathbb{E}[X_i]$. La **media muestral** se define como:

$$
\bar{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

La Ley de los Grandes Números establece que, cuando $n \to \infty$:

$$
\bar{X}_n \xrightarrow{a.s.} \mu
$$

donde $\xrightarrow{a.s.}$ significa **convergencia casi segura**.

**Interpretación:**  
A medida que aumentamos el número de observaciones, la media muestral se aproxima cada vez más a la media verdadera de la distribución.

---

## Tipos de LGN

1. **Ley Débil de los Grandes Números (LDGN):**  
$$
\bar{X}_n \xrightarrow{P} \mu
$$  
Convergencia en probabilidad, es decir, para cualquier $\epsilon > 0$:
$$
\lim_{n \to \infty} P(|\bar{X}_n - \mu| > \epsilon) = 0
$$

2. **Ley Fuerte de los Grandes Números (LFGN):**  
$$
\bar{X}_n \xrightarrow{a.s.} \mu
$$  
Convergencia casi segura, un resultado más fuerte que la ley débil.

---

## Ejemplo conceptual

Si lanzamos un dado justo muchas veces:

- $X_i$ = resultado del $i$-ésimo lanzamiento
- $\mu = 3.5$ (media del dado)
- A medida que el número de lanzamientos aumenta, la media de los resultados observados se aproxima a $3.5$


```python
import numpy as np
import matplotlib.pyplot as plt

# Número de simulaciones
n = 10000

# Simulación: lanzamientos de un dado justo (1-6)
X = np.random.randint(1, 7, size=n)

# Media acumulada
mean_cum = np.cumsum(X) / np.arange(1, n+1)

# Visualización
plt.figure(figsize=(10,5))
plt.plot(mean_cum, color='blue', lw=2)
plt.axhline(y=3.5, color='red', linestyle='--', label='Media teórica')
plt.title("Ley de los Grandes Números: Media acumulada de lanzamientos de un dado")
plt.xlabel("Número de lanzamientos")
plt.ylabel("Media acumulada")
plt.legend()
plt.show()
```
